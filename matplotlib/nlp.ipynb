```python
import nltk as nlp
from nltk.corpus import stopwords, state_union
from nltk.stem import PorterStemmer
nlp.download('punkt')
nlp.download('stopwords')
```

```python
#tokenizing-- word tokenizer and sentence tokenizer
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
#print the splitted sentences
ex="Hello there, how are you? How are you doning today. Weather is good and python is best."
print(sent_tokenize(ex))
```

```python
#print the splitted words
print(word_tokenize(ex))
```

```python
#STOP WORDS
# a, an, oh, these kind of words are called stop words
# we want to remove that kind of words
ex2='this is an example showing of stopword filteration.'
stop_words=set(stopwords.words("english"))
words=word_tokenize(ex2)
fs=[]

for w in words:
  if w not in stop_words:
    fs.append(w)
```

```python
print(fs)
```

```python
      #STEMING IS IMPORTANT

ps=PorterStemmer()
explewords=["run","running","ran","ran"]
for w in explewords:
  print(ps.stem(w))
```
