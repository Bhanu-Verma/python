```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
        #SCATTER PLOTS

arr=[9.1,9.2,9.3,9.4,9.5,9.6,9.7,9.8,9.9]
y=[]
x=[]
i=0
while(i<50):
  y.append(random.choice(arr))
  i+=1
#print(y)
for i in range(50):
  x.append(i)
#print(x)
plt.xlabel("index")
plt.ylabel("value generated")
plt.title("checking the randomness of random")
plt.scatter(x,y,color="hotpink",marker="*")
plt.legend
```

```python
        #STACK PLOTS
days=[1,2,3,4,5]
sleep=[6,7,5,10,12]
working=[7,8,2,10,12]
playing=[2,4,1,5,6]

plt.stackplot(days, sleep,working,playing,colors=['r','b','m'])
plt.show()
plt.xlabel("index")
plt.ylabel("value generated")
plt.title("checking the randomness of random")
plt.scatter(x,y,color="hotpink",marker="*")
```

```python
      #PIE CHART

#as default pie chart starts from the horizontal axis after it all the elements are represented in the anticlockwise order
#to start this with any other angle we can use the attribute startangle

#attributes:
#startangle: to set the angle where the chart starts from
#colors: to specify the colors. Just pass in the form of a list
#shadow: if we write shadow=true it will add little shadow to the chart
#explode: it is used if we want our one of the elements to be in the form of a pizza slice 
times=[7,2,3,8,4]
activity=['sleeping','eating','rest','reading','stuff']

plt.pie(times,
        labels=activity,
        colors=['red','green','blue','violet','yellow'],
        shadow=True,
        explode=[0,0,0,0.2,0],
        
        )
plt.show()
```
