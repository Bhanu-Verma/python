```python
    #PLOTING OF BAR CHARTS
x=[2,4,6,8,10]
y=[4,16,36,64,100]
plt.xlabel("number")
plt.ylabel("square")
plt.title("graph of the squares\nit's easy but worth looking")
plt.bar(x,y,label="bars1",color="green",width=1)
plt.legend()
```
