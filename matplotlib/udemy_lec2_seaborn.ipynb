```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ourdata = pd.read_csv("running.csv")
ourdata
```

```python
#plot for a linear regression model--

sns.lmplot(x='Climb (m)',y='Average Heart Rate (bpm)',data=ourdata)
```

```python
#to ger rid of the regression line
sns.lmplot(x='Climb (m)',y='Average Heart Rate (bpm)',data=ourdata,fit_reg=False)
```

```python
sns.lmplot(x='Climb (m)',y='Average Heart Rate (bpm)',data=ourdata,fit_reg=False,hue='Calories Burned')
```
