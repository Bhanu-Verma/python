```python
#now we will learn a method to do all this stuff while loeading the data only
def d_parse(x):
  pd.to_datetime(x,format="%Y-%m-%d %H:%M:%S")
  return x

df2=pd.read_csv('/content/dates.csv',parse_dates=['Date'], date_parser=d_parse)

#first argument of parse_dates is the list of columnname of all the columns that we want to parse
#then we srite date_parser and set that equal to the function that we want to use for parsing the dtaes
```

```python
df2.info()    #now we are getting it as datetime format only
```

```python
#if we want to know the name of all date column then we can do like it--
#NAIVE  APPROACH
for date in df2['Date']:
  print(date.day_name())
```

```python
#EFFICIENT SOLUTION
df2['Date'].dt.day_name()   #dt is like str from a series to find all the date and time we use this
```

```python
#to access the earliest and last date of a data frame
df2['Date'].min()
```

```python
df2['Date'].max()
```

```python
#to get the amount of time in which a data is taken
df2['Date'].max()-df2['Date'].min()

```

```python
#FILTERING THE DATA ON THE BASIS OF DATES
filt = (df2['Date'] > '2017')
```

```python
df2.loc[filt]   #this will only give dates after 2017
```

```python
#to do it more precise we use pd.to_datetime method
#we'll filter all the dates of november from the dataset
filt = ((df2['Date']> pd.to_datetime('2018-10-31')) & (df2['Date']< pd.to_datetime('2018-12-01')))
df2[filt]
```

```python
#in this kind of data it is good to set date as the index of df
df2.set_index("Date",inplace=True)
```

```python
#now for accessing data of 2018
df2['2018']
```

```python
#to get the data for nov
df2['2018-11-01':'2018-11-30']
```

```python
#now we can easily find the mean of distance that he ran in november
df2['2018-11-01':'2018-11-30']['Distance (km)'].mean()
```

```python
#RESAMPLING
#suppose our data is distributed on the hourly basis or on monthly basis
#we want to change its distribution period for it we can use resampling

#this piece of code gives the max value of distance on monthly basis
#in resample funcion we have a lot of argument values M is for month
df2['Distance (km)'].resample('M').max()
```

```python
 #here we resample only one column but we can use this method on the complete dataframe also
 df2.resample('M').mean()
```

```python
#now suppose we want to find total distacnce, total calories burned, but avg speed, avg heart rate
df2.resample('M').agg({'Distance (km)':'sum','Average Speed (km/h)':'mean','Calories Burned':'mean',})
```
