Join And Merge


```python
# Dataframe of number of sales made by an employees
import numpy as np
sales = {'Jones': 10000,
         'Chris': 5000,
         'Piyush': 440,
         'Meera': 6700,
         'Rahul': 300
        }
# Dataframe of all employees and the region they work in
region = {'Jones': 'West',
          'Chris': np.nan,
          'Piyush': 'West',
          'Meera': np.nan,
          'Anthony': 'East',
          'Ellen': 'South',
          'Josh': 'West',
          'Simran': 'East',
          'Oscar': 'North',
         }
```

```python
sales_df = pd.DataFrame.from_dict(sales,orient='index',columns=['sales'])
region_df = pd.DataFrame.from_dict(region,orient='index',columns=['region'])

#orient has three arguments--
#'index' = if you want to use keys as row index
#'columns' = if you want to use keys as column index(default)
```

```python
sales_df
```

```python
region_df
```

```python
extra_df = pd.DataFrame(sales,index=range(1))
extra_df
```

JOINING the data frames---
left--only those things that are in left df
right--only those things that are in right df
inner--things present in their intersection
outer--union will come

```python
joined_df1 = region_df.join(sales_df,how='left')
joined_df1      #see rahul has disappeared
```

```python
joined_df2 = region_df.join(sales_df,how='right')
joined_df2  #only those keys that are present in sales_df
```

```python
joined_df3 = region_df.join(sales_df,how='inner')
joined_df3      #only intersection is coming
```

```python
joined_df4 = region_df.join(sales_df,how='outer')
joined_df4      #union will appear
```

Merge--for merging we need to specify on which column we are going to merge so we'll give our index column a name
* how works here as well 


```python
#giving title to the index column--
sales_df.index.name = 'names'
region_df.index.name = 'names'
sales_df
```

```python
merged_df = pd.merge(region_df,sales_df,on='names')

merged_df
```

```python
merged_df = pd.merge(region_df,sales_df,on='names',how='left')

merged_df
```

```python
merged_df = pd.merge(region_df,sales_df,on='names',how='right')

merged_df
```
