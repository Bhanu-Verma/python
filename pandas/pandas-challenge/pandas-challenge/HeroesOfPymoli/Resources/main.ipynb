```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('purchase_data-checkpoint.csv',index_col='Purchase ID')
df
```

```python
unique_players = len(df['SN'].unique())
print(f"total number of palyers is {unique_players}")
```

```python
### Purchasing Analysis (Total)
unique_items = len(df['Item ID'].unique())
print(f"total number of unique items is {unique_items}")
avg_price = df['Price'].mean()
print(f"average purchase price is ${avg_price}")
total_purchase = df['Item ID'].count()
print(f"total purchases are {total_purchase}")
total_revenue = df['Price'].sum()
print(f"total revenue is ${total_revenue}")
```

```python
### Gender Demographics
male_df = df.groupby(['Gender']).get_group('Male')
males = len(male_df['SN'].unique())
male_percente = round(males/unique_players*100,2)

female_df = df.groupby(['Gender']).get_group('Female')
females = len(female_df['SN'].unique())
female_percente = round(females/unique_players*100,2)

others_df = df.groupby(['Gender']).get_group('Other / Non-Disclosed')
others = len(others_df['SN'].unique())
others_percente = round(others/unique_players*100,2)

gender_df = pd.DataFrame(
    {'Gender':['Male','Female','Other / Non-Disclosed'],
     'Count':[males,females,others],
     'Percentage':[male_percente,female_percente,others_percente]
     }
)
gender_df
```

```python
### Purchasing Analysis (Gender)
gender_purchases_df=df.groupby(['Gender'])['Item Name'].count()
gender_purchases_df


average_purchases_df=round(df.groupby(['Gender'])['Price'].mean(),2)
average_purchases_df


total_purchases_df=round(df.groupby(['Gender'])['Price'].sum(),2)
total_purchases_df

avg_total_purchases_df=round(total_purchases_df/gender_purchases_df,2)
avg_total_purchases_df

gender_analysis_df = pd.DataFrame({
    'Purchase Count':gender_purchases_df,
    'Average Purchase price':average_purchases_df,
    'Total purchase Price':total_purchases_df,
    'Avg_price_per_person':avg_total_purchases_df
})

gender_analysis_df
```
