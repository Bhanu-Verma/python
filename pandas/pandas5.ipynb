```python
#QUE--we want to see in which country which social media platform is more famous
#for it we have two columns socialmedia and country
#THIS IS WHERE WE FEEL THE NEED OF GROUPING
#grouping can be divided in three processes
#---split
#----applyfunction
#----join
df['Country'].value_counts()
```

```python
country_grp = df.groupby(['Country'])
```

```python
country_grp.get_group('India')    #gives all the rows with country as Indai
```

```python
country_grp['SocialMedia'].value_counts()   #it is similiar to 
                                            #applying a filter 
                                            #of country and 
                                            #then get values_counts
                                            #but then it will take a 
                                            #lot of time
```

```python
#Now suppose we want this only for India
country_grp['SocialMedia'].value_counts().loc['India']
```

```python
#now let's try to find median salary for different countries
country_grp['ConvertedComp'].median().sort_values(ascending=False)
```

```python
#to find median salary in India
country_grp['ConvertedComp'].median().loc['India']
```

```python
#AGG FUNCTION
#IF WE WANT TO SEE BOTH MEDIAN AND MEAN THEN WE CAN APPLY BOTH OF THEM SIMULTANEOUSLY USING AGG
country_grp['ConvertedComp'].agg(['median','mean']).loc['India']
```
