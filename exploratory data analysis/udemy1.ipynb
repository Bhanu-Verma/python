Understand Your Data With Visualization

```python
#univariate plots---
# 1.histogram
# 2.density plot
# 3.box and whisker plot
```

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
df.plot(kind='density', subplots =True, layout=(3,3),sharex=False,sharey=False, figsize=(15,10))
plt.show()
```

```python
df.plot(kind='box',layout=(3,3),subplots=True,sharex=False,sharey=False,figsize=(12,12))
plt.show()
#Here we can see some of the candidate doutliers
```

```python
#1.correlation matrix plot---
import seaborn as sns
plt.figure(figsize=(6,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```
